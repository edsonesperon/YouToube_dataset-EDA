Como parte del preprocesamiento, eliminaremos valores que no son de interés para este problema, como lo son las variables no númericas y los valores negativos.

```python
#Eliminando variables no numéricas
df=df.select_dtypes(exclude=['object'])
```

```python
#Eliminando valores negativos
df2=df[(df>=0).all(1)]
```

Creamos un nuevo dataframe (`df2`) con las modificaciones que se consideraron pertinentes.

```python
df2
```

Una vez terminado un preprocesamiento sencillo, procedemos a estandarizar los datos y a hacer un _heatmap_ de correlaciones para hallar la variable más correlacionada con `likes/views`. 

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled_df=ss.fit_transform(df2)
scaled_df=pd.DataFrame(scaled_df, columns=df2.columns)
```

```python
f, ax=plt.subplots(figsize=(10,6))
#utilizando la función corr() de pandas
corr=scaled_df.corr()
#heatmap se utiliza para graficar la matriz de correlaciones
hm=sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm", fmt=".2f",
               linewidths=.05)
f.subplots_adjust(top=0.93)
t=f.suptitle("Correlation Heatmap", fontsize=14)
```

De primera mano no podemos asegurar cuál es la variable que está más correlacionada con `likes/views`.

Por lo tanto, tomamos a la matriz de correlaciones (corr) y se aplica valor absoluto porque la correlación puede ser positiva o negativa. Posteriormente, las ordenamos de menor a mayor.

```python
#Correlaciones más altas en valor absoluto
abs(corr['likes/views']).sort_values()
```

Con lo anterior podemos notar que la variable que está más correlacionada con `likes/views` es `likes/dislikes`.

Ahora, procederemos a hacer unas gráficas para poder obtener información relevante de esta correlación entre variables.

```python
df2['likes/views'].hist(bins=15, color="green", edgecolor="black", linewidth=1.0,
                        xlabelsize=8, ylabelsize=8, grid=True)
plt.tight_layout(rect=(0,0,1.2,1.2))
```

```python
jp=sns.jointplot(x="likes/views", y="likes/dislikes", data=scaled_df, kind="reg", space=0, height=6, ratio=4)
```

De las gráficas podemos concluir que para monetizar más analizando la variable `likes/views` tenemos que:
* Aumentar la cantidad de suscriptores.
* Aumentar la cantidad de comentarios en los videos. 
* Disminuir la cantidad de dislikes.
